```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
baseline_df = pd.read_csv('baseline_results.csv')
pipeline_df = pd.read_csv('pipeline_results.csv')
```

```python
display(baseline_df.head())
display(pipeline_df.head())
```

```python
# Compute averages for Flesch-Kincaid Grade Level and Reading Ease
avg_fk_baseline = baseline_df['FK_Grade'].mean()
avg_fre_baseline = baseline_df['FRE_Score'].mean()

avg_fk_pipeline = pipeline_df['FK_Grade'].mean()
avg_fre_pipeline = pipeline_df['FRE_Score'].mean()

print("Baseline - FK Grade:", avg_fk_baseline, "FRE Score:", avg_fre_baseline)
print("Pipeline - FK Grade:", avg_fk_pipeline, "FRE Score:", avg_fre_pipeline)
```

```python
labels = ['Flesch-Kincaid Grade', 'Flesch Reading Ease']
baseline_values = [avg_fk_baseline, avg_fre_baseline]
pipeline_values = [avg_fk_pipeline, avg_fre_pipeline]

# Bar chart
x = range(len(labels))
plt.bar(x, baseline_values, width=0.4, label='Baseline', align='center')
plt.bar([xi + 0.4 for xi in x], pipeline_values, width=0.4, label='Pipeline', align='center')

plt.xticks([xi + 0.2 for xi in x], labels)
plt.ylabel('Score')
plt.title('Comparison of Readability Metrics')
plt.legend()
plt.savefig('comparison_chart.png')  # Save the chart as a PNG file
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(baseline_df['FK_Grade'], label='Baseline FK Grade', marker='o')
plt.plot(pipeline_df['FK_Grade'], label='Pipeline FK Grade', marker='o')
plt.plot(baseline_df['FRE_Score'], label='Baseline FRE Score', marker='o')
plt.plot(pipeline_df['FRE_Score'], label='Pipeline FRE Score', marker='o')

plt.xlabel('Question Index')
plt.ylabel('Score')
plt.title('Readability Metrics for Each Question')
plt.legend()
plt.savefig('detailed_comparison.png')  # Save as PNG
plt.show()
```
